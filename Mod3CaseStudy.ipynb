I thought about using @dataclass which I learned about in this week's reading because the app is concerned with storing and presenting data. I think it would be more pythonic. But to be sure I fullfilled the assignment I am using this form with lots of boilerplate.

```python
class Vehicle:
#Super class to store the type of vehicle.
    def __init__(self, vehicle_type):
        self.vehicle_type = vehicle_type


class Automobile(Vehicle):
#Class to store details of a car.
    def __init__(self, vehicle_type, year, make, model, doors, roof):
        # Initialize super class attribute
        super().__init__(vehicle_type)
        # Initialize Automobile attributes
        self.year = year
        self.make = make
        self.model = model
        self.doors = doors
        self.roof = roof

    def display_info(self):
    #Print vehicle information in an easy to read format
        print("\n  Vehicle Details  \n")
        print(f"Vehicle type:   {self.vehicle_type}")
        print(f"Year:           {self.year}")
        print(f"Make:           {self.make}")
        print(f"Model:          {self.model}")
        print(f"Number of doors:{self.doors}")
        print(f"Type of roof:   {self.roof}")
```

```python
#The logic of the app
def main():
    print("Vehicle Registration App\n")
 
    vehicle_type = "car"

    # Collect user input
    year = input("Enter the year: ")
    make = input("Enter the make: ")
    model = input("Enter the model: ")
    
    # Input validation for doors
    while True:
        doors = input("Enter the number of doors (2 or 4): ")
        if doors in ["2", "4"]:
            break
        print("Invalid input. Please enter 2 or 4.")

    # Input validation for roof
    while True:
        roof = input("Enter the type of roof (solid or sun roof): ").lower()
        if roof in ["solid", "sun roof", "sunroof"]:
            break
        print("Invalid input. Please enter 'solid' or 'sun roof'.")

    # Creating the Automobile object
    user_car = Automobile(vehicle_type, year, make, model, doors, roof)

    # Output data collected using the display function
    user_car.display_info()
    
```

```python
#Call the main function for execution
if __name__ == "__main__":
    main()
```

The app has worked as intended.
