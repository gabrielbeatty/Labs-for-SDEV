#Required prior to 16.8

```python
import sqlite3

# Connect to the database file
# Will be in the same folder as the notebook itself
conn = sqlite3.connect('books.db')
cursor = conn.cursor()

# Create the table
cursor.execute('''
    CREATE TABLE IF NOT EXISTS books (
        title TEXT,
        author TEXT,
        year INTEGER
    )
''')

# data set needed 
books_to_add = [
    ('Inferno', 'Dante Alighieri', 1321),
    ('Angels', 'Billy Graham', 1321),
]
cursor.executemany('INSERT INTO books VALUES (?,?,?)', books_to_add)

# Save and close
conn.commit()
conn.close()
```

#16.8

```python
from sqlalchemy import create_engine, text

# 'sqlite:///' (3 slashes) indicates a relative path to this file
engine = create_engine('sqlite:///books.db')

with engine.connect() as connection:
  
    query = text("SELECT title FROM books ORDER BY title ASC")
    result = connection.execute(query)
    
    print("Titles in Alphabetical Order:")
    for row in result:       
        print(f"- {row.title}")
```
