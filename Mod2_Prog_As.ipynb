##6.2

```python
guess_me = 7
number = 1

while True:
    if guess_me < number:
        print("too low")
        break
    elif guess_me == number:
        print("found it!")
        break
    elif guess_me > number:
        print("oops")
        break
    
```

##6.3

```python
guess_me = 5

for number in range(1,11,1):
    if number < guess_me:
        print("too low")
        continue
    elif number == guess_me:
        print("found it!")
        break
    elif number > guess_me:
        print("oops")
        continue
```

for this code I chose to use break once the number was identified.
