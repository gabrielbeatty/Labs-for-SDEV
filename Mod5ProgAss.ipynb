```python
def sum(arg):
    total = 0
    for val in arg:
        total += val
    return total
```

```python
import unittest
from fractions import Fraction

# The test class containing both the integer test and the fraction test
class TestSum(unittest.TestCase):
    def test_list_int(self):
        
        #Test that it can sum a list of integers
        
        data = [1, 2, 3]
        result = sum(data)
        self.assertEqual(result, 6)

    def test_list_fraction(self):
        
        #Test that it can sum a list of fractions
        
        data = [Fraction(1, 4), Fraction(1, 4), Fraction(2, 5)]
        result = sum(data)
        # 1/4 + 1/4 + 2/5 = 0.9 (this test is intended to fail!)
        self.assertEqual(result, 1)

if __name__ == '__main__':
    # Using 'verbosity=2' will show you the name of each test as it runs
    unittest.main(argv=['first-arg-is-ignored'], verbosity=2, exit=False)
```

The first test passed because the sum of the given data set is indeed equal to 6.(The assertion.)
The second test failed because the assertion was not true and the sum of the data set was not equal to 1.
