Sort an Array of 0's, 1's, and 2's

```python
class Solution:
    def sort012(self, arr):
        
        # Initialize three pointers
        low = 0
        mid = 0
        high = len(arr) - 1

        # Pass through the array
        while mid <= high:
            if arr[mid] == 0:
                # Swap current with low pointer 
                arr[low], arr[mid] = arr[mid], arr[low]
                low += 1
                mid += 1
            elif arr[mid] == 1:
                # 1 is in the middle, move forward
                mid += 1
            else: # arr[mid] == 2
                # Swap current with high pointer
                arr[mid], arr[high] = arr[high], arr[mid]
                high -= 1
        
        return arr
# Example array
nums = [2, 0, 2, 1, 1, 0]

# Instance of the class
sol = Solution()        
print(sol.sort012(nums)) 
```

Binary Search

```python
class Solution:
    def binarysearch(self, arr, k):
        low = 0
        high = len(arr) - 1
        result = -1
        
        while low <= high:
            # Use floor division // for the middle index
            mid = low + (high - low) // 2
            
            if arr[mid] == k:
                result = mid 
                # Store index but keep looking left for any smaller index that matches
                high = mid - 1    
            elif arr[mid] < k:
                low = mid + 1
            else:
                high = mid - 1
                
        return result

sol = Solution()
print(sol.binarysearch([1, 1, 1, 1, 2], 1))
# Outputs 0 (the smallest index where k is found)
print(sol.binarysearch([1, 2, 3, 4, 5], 4)) 
```
